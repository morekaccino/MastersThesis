# Signal Processing Methodology for Ion Channel Analysis

## Introduction
The analysis of ion channel behavior is a crucial area of study in biology, as it provides insights into fundamental cellular processes and plays a significant role in various physiological functions and diseases. However, experimental data obtained from ion channel recordings often presents challenges due to inherent noise and uncertainties. In this research, we developed a comprehensive signal processing methodology to analyze noisy digital signals representing ion channel activity, with the goal of estimating the number of underlying states and characterizing their properties.

## Data Acquisition and Preprocessing
We obtained noisy digital signals representing the behavior of ion channels from biological experimental data. These signals exhibit characteristics similar to digital waveforms but with added noise and variability due to the stochastic nature of ion channel dynamics and experimental conditions.

To ensure computational efficiency and reduce the impact of noise, we preprocessed the raw signal data through downsampling or filtering techniques, if necessary. The preprocessing step aimed to retain the essential signal information while mitigating the effects of high-frequency noise or irrelevant components.

## Data Visualization
Prior to applying signal processing algorithms, we visualized the preprocessed signal data using scatter plots or line plots. This visualization step allowed us to gain insights into the signal's characteristics, such as potential state levels, noise levels, and overall complexity. It also helped inform the selection and parameterization of subsequent analysis techniques.

## State Estimation and Clustering
To estimate the number of underlying states present in the ion channel signal and classify individual data points into their respective states, we employed three complementary approaches: histogram-based analysis, Gaussian Mixture Models (GMMs), and Bayesian Gaussian Mixture Models (BGMMs).

### Histogram-based Approach
In the histogram-based approach, we binned the signal data into a specified number of bins and computed the histogram to obtain the distribution of signal values. To mitigate the effects of noise, we applied a moving average filter to the histogram, effectively smoothing the distribution.

We then identified peaks in the smoothed histogram that exceeded a predefined threshold, with the number of significant peaks serving as an estimate of the number of states or clusters present in the signal.

To further refine the state estimation, we performed DBSCAN clustering on the peak coordinates, allowing us to identify potential outliers and obtain a more robust estimate of the number of clusters.

### Gaussian Mixture Model (GMM) Approach
The GMM approach involved fitting a Gaussian Mixture Model to the signal data, with varying numbers of components. We evaluated the model fit using the Bayesian Information Criterion (BIC) or other information-theoretic criteria and selected the optimal number of components based on the minimum BIC score or other criteria.

The means, standard deviations, and weights of the fitted Gaussian distributions for the best GMM provided valuable insights into the characteristics of the identified states, such as their central tendencies, spread, and relative prevalence.

We visualized the BIC scores for different numbers of components to assess the model selection process and ensure the chosen number of states was appropriate.

### Bayesian Gaussian Mixture Model (BGMM) Approach
Building upon the results from the histogram-based and GMM approaches, we fitted a Bayesian Gaussian Mixture Model (BGMM) to the signal data using the estimated number of clusters or states.

The BGMM allowed us to classify individual data points into their most probable state or cluster based on the model predictions. We computed the confidence scores or posterior probabilities associated with these classifications, providing insights into the certainty of the state assignments.

To facilitate interpretation and analysis, we visualized the classified signal data, with the point transparency or color representing the confidence of the classification.

## Find the number of states

```python
# df.signal find the bins distribution

bins = np.linspace(df.signal.min(), df.signal.max(), 200)
np.digitize(df.signal, bins)
# plot the bins
hists = plt.hist(df.signal, bins=bins, alpha=0.5, color='b')

# find the moving average of hists
ma = moving_average(hists[0], 5)
ma = moving_average(ma, 5)
plt.plot(hists[1][:-1], ma, 'r', linewidth=5, label='moving average')

# find number of peaks in the moving average that are above a certain threshold
threshold = 0
peaks = np.diff(np.sign(np.diff(ma))) < 0
peaks = np.where(peaks)[0] + 1
number_of_states = sum(ma[peaks] > threshold)
plt.plot(hists[1][peaks][ma[peaks] > threshold], ma[peaks][ma[peaks] > threshold], 'o', markersize=10, label='peaks',
         color='g')
plt.legend()
plt.show()

print(f'Number of states: {number_of_states}')
```

```python
# DBSCAN on hists[1][peaks], ma[peaks]
peaks_coords = np.array(list(zip(hists[1][peaks], ma[peaks])))
# remove peaks that are less than the threshold
threshold = 5
peaks_coords = peaks_coords[ma[peaks] > threshold]

dbscan = DBSCAN(eps=10, min_samples=1)
dbscan.fit(peaks_coords)
labels = dbscan.labels_
n_clusters_ = len(set(labels)) # - (1 if -1 in labels else 0)
print(f'Number of clusters: {n_clusters_}')

# plot the clusters
plt.scatter(peaks_coords[:, 0], peaks_coords[:, 1], c=labels, cmap='viridis', s=50, alpha=1, marker='o')
plt.plot(hists[1][:-1], ma, 'y', linewidth=1, label='moving average', alpha=1)
plt.legend()
labels
```

```python
lowest_bic = np.infty
lowest_aic = np.infty
best_gmm_bic = None
best_gmm_aic = None

n_components_range = range(1, 10)
bic_scores = []
for n_components in n_components_range:
    gmm = GaussianMixture(n_components=n_components, covariance_type='full')
    gmm.fit(df.signal.values.reshape(-1, 1))
    bic_scores.append(gmm.bic(df.signal.values.reshape(-1, 1)))

# Select the best model based on the lowest BIC score
best_model_index = np.argmin(bic_scores)
best_n_components = n_components_range[best_model_index]
best_gmm = GaussianMixture(n_components=best_n_components, covariance_type='full')
best_gmm.fit(df.signal.values.reshape(-1, 1))

# Print the means, standard deviations, and weights of the fitted normal distributions
means = best_gmm.means_.flatten()
stds = np.sqrt(best_gmm.covariances_).flatten()
weights = best_gmm.weights_

# plot the BIC scores
fig, ax = plt.subplots(1,2)
ax[0].plot(n_components_range, bic_scores, marker='o')
ax[0].set_xlabel('Number of components')
ax[0].set_ylabel('BIC score')
plt.show()

# plot real signal distributions in ax[1]
df.plot(x='time', y='signal', figsize=(30, 5), kind='scatter', alpha=0.5, c='b', ax=ax[1])




print("Number of components:", best_n_components)
print("Means:", means)
print("Standard deviations:", stds)
print("Weights:", weights)
```
