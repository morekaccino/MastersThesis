```python
import numpy as np
import pandas as pd
from source.moreka import AxonData
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import euclidean_distances
from scipy.spatial.distance import pdist, squareform
import matplotlib.pyplot as plt
```

```python
data = AxonData(dirname='./data')
df = data[3]#.iloc[::50, :]

df['difference'] = abs(df['signal'].diff().fillna(0))
df
```

```python
df.plot(x='time', y='difference', kind='scatter', alpha=0.3, s=5)
df.plot(x='time', y='signal', kind='scatter', alpha=0.3, s=5)
```

```python
diff = np.array(df.difference)
stds = []
means = []
window_size = 100
for i in range(0, len(diff), window_size):
    stds.append(diff[i:i+window_size].std())
    means.append(diff[i:i+window_size].mean())
```

```python
plt.plot(range(len(stds)), stds)
```

```python
plt.plot(range(len(stds)), means)
```
